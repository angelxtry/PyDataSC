## 7.3.1 중복 제거하기

```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
```

```python
data = DataFrame({'k1': ['one']*3 + ['two']*4,
                  'k2': [1, 1, 2, 3, 3, 4, 4]})
```

```python
data
```

* DataFrame의 duplicated 메서드는 각 로우가 중목인지 아닌지를 알려주는 불리언 Series 객체를 반환한다.

```python
data.duplicated()
```

* drop_duplicates는 duplicated 배열이 False인 DataFrame을 반환한다.

```python
data.drop_duplicates()
```

* 특정 칼럼을 지정하여 해당 칼럼에 기반해서 중복을 걸러낼 수도 있다.

```python
data['v1'] = range(7)
```

```python
data
```

```python
data.drop_duplicates(['k1'])
```

* duplicated와 drop_duplicates는 기본적으로 처음 발결된 값을 유지한다.
* keep='last' 옵션을 넘기면 마지막으로 발견된 값을 반환한다.

```python
data.drop_duplicates(['k1', 'k2'], keep='last')
```

## 7.3.2 함수나 매핑 이용해 데이터 변형하기

```python
data = DataFrame({'food': ['bacon', 'pulled pork', 'bacon', 'Pastrami',
                           'corned beef', 'Bacon', 'pastrami',
                           'honey ham', 'nova lox'],
                  'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
```

```python
data
```

```python
meat_to_animal = {
    'bacon': 'pig',
    'pulled pork': 'pig',
    'pastrami': 'cow',
    'corned beef': 'cow',
    'honey ham': 'pig',
    'nova lox': 'salmon',
}
```

* Series의 map 메서드는 사전류의 객체나 함수를 인자로 받을 수 있다.

```python
data['animal'] = data['food'].map(str.lower).map(meat_to_animal)
```

```python
data
```

```python
data['food'].map(lambda  x: meat_to_animal[x.lower()])
```

## 7.3.4 축 색인 이름 바꾸기

```python
data = DataFrame(np.arange(12).reshape((3, 4)),
                 index=['Ohio', 'Colorado', 'New York'],
                 columns=['one', 'two', 'three', 'four'])
```

```python
data
```

* Series와 마찬가지로 축 색인에도 map 메서드가 있다.

```python
data.index.map(str.upper)
```

* 변경된 축 이름은 DataFrame의 index에 바로 대입할 수 있다.

```python
data.index = data.index.map(str.upper)
```

```python
data
```

* rename 메서드를 사용하면 원래 객체를 변경하지 않고 새로운 객체를 생성한다.

```python
data.rename(index=str.title, columns=str.upper)
```

* rename 메서드는 사전 형식의 객체를 이용해서 축 이름 중 일부만 변경하는 것도 가능하다.

```python
data.rename(index={'OHIO': 'INDIANA'},
            columns={'three': 'peekaboo'})
```

```python
_ = data.rename(index={'OHIO': 'INDIANA'}, inplace=True)
```

## 7.3.5 개별화와 양자화

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

```python
bins = [18, 25, 35, 60, 100]
```

* cut에서 반환하는 객체는 Categorical 객체라는 특수한 객체다.
* 그룹 이름이 담긴 배열이라고 생각하면 된다.

```python
cats = pd.cut(ages, bins)
```

```python
cats
```

```python
cats.codes
```

```python
cats.categories
```

```python
pd.value_counts(cats)
```

* 간격을 나타내는 표기법은 괄호로 시작해서 대괄호로 끝난다.
* 괄호 쪽의 값은 포함하지 않고 대괄호 쪽의 값은 포함하는 간격을 나타낸다.
* right=False를 넘겨서 괄호와 대괄호의 위치를 바꿀 수 있다.

```python
pd.cut(ages, [18, 26, 36, 61, 100], right=False)
```

* labels 옵션으로 그룹의 이름을 직접 넘겨줄 수도 있다.

```python
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
```

```python
pd.cut(ages, bins, labels=group_names)
```

* cut 

```python
data = np.random.rand(20)
```

## 7.3.6 특이값 찾아내고 제외하기

* 배열연산을 수행할 때는 특이값 outlier을 제외하거나 적당한 값으로 대체하는 것이 중요하다.

```python
np.random.seed(12345)
```

```python
data = DataFrame(np.random.randn(1000, 4))
```

```python
data.describe()
```

```python
col = data[3]
```

```python
col[np.abs(col) > 3]
```

```python
data[(np.abs(data) > 3).any(1)]
```

```python
data[np.abs(data) > 3] = np.sign(data) * 3
```

```python
# np.sign return -1 if x < 0, 0 if x==0, 1 if x > 0
```

```python
data.describe()
```
