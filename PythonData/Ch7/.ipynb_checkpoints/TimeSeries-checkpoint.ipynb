```python
pinkwink_web = pd.read_csv('../data/08.PinkWink_Web_Traffic.csv',
                           encoding='utf-8', thousands=',',
                           names=['date', 'hit'], index_col=0)
```

```python
pinkwink_web = pinkwink_web[pinkwink_web['hit'].notnull()]
```

```python
pinkwink_web.head(10)
```

```python
pinkwink_web['hit'].plot(figsize=(12, 4), grid=True);
```

```python
time = np.arange(0, len(pinkwink_web))
```

```python
traffic = pinkwink_web['hit'].values
```

```python
fx = np.linspace(0, time[-1], 1000)
```

```python
def error(f, x, y):
    return np.sqrt(np.mean((f(x)-y)**2))
```

```python
fp1 = np.polyfit(time, traffic, 1)
f1 = np.poly1d(fp1)
```

```python
f2p = np.polyfit(time, traffic, 2)
f2 = np.poly1d(f2p)
```

```python
f3p = np.polyfit(time, traffic, 3)
f3 = np.poly1d(f3p)
```

```python
f15p = np.polyfit(time, traffic, 15)
f15 = np.poly1d(f15p)
```

```python
print(error(f1, time, traffic))
print(error(f2, time, traffic))
print(error(f3, time, traffic))
print(error(f15, time, traffic))
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(time, traffic, s=10)

plt.plot(fx, f1(fx), lw=4, label='f1')
plt.plot(fx, f2(fx), lw=4, label='f2')
plt.plot(fx, f3(fx), lw=4, label='f3')
plt.plot(fx, f15(fx), lw=4, label='f15')

plt.grid(True, linestyle='-', color='0.75')
plt.legend(loc=2)
plt.show()
```

```python
df = pd.DataFrame({'ds': pinkwink_web.index, 'y': pinkwink_web['hit']})
```

```python
df.reset_index(inplace=True)
```

```python
df['ds'] = pd.to_datetime(df['ds'], format="%y. %m. %d.")
```

```python
del df['date']
```

```python
m = Prophet(yearly_seasonality=True)
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=60)
```

```python
future.tail(10)
```

```python
forecast = m.predict(future)
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast)
```

```python
m.plot_components(forecast)
```

```python
df = pd.read_csv('../data/08.example_wp_peyton_manning.csv')
df['y'] = np.log(df['y'])
m = Prophet(daily_seasonality=True)
m.fit(df)
future = m.make_future_dataframe(periods=366)
```

```python
df.y.plot(figsize=(12,6), grid=True);
```

```python
playoffs = pd.DataFrame({
  'holiday': 'playoff',
  'ds': pd.to_datetime(['2008-01-13', '2009-01-03', '2010-01-16',
                        '2010-01-24', '2010-02-07', '2011-01-08',
                        '2013-01-12', '2014-01-12', '2014-01-19',
                        '2014-02-02', '2015-01-11', '2016-01-17',
                        '2016-01-24', '2016-02-07']),
  'lower_window': 0,
  'upper_window': 1,
})
superbowls = pd.DataFrame({
  'holiday': 'superbowl',
  'ds': pd.to_datetime(['2010-02-07', '2014-02-02', '2016-02-07']),
  'lower_window': 0,
  'upper_window': 1,
})
holidays = pd.concat((playoffs, superbowls))
```

```python
m = Prophet(holidays=holidays, daily_seasonality=True)
forecast = m.fit(df).predict(future)
```

```python
forecast[(forecast['playoff'] + forecast['superbowl']).abs() > 0][
        ['ds', 'playoff', 'superbowl']][-10:]
```

```python
m.plot(forecast);
```

```python
m.plot_components(forecast);
```
