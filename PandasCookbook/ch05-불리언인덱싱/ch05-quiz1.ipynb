## 불리언 통계량

1.영화 데이터셋을 로드하여 상영 시간(duration)이 2시간이 넘는 영화의 개수를 확인하세요.

```python
movie = pd.read_csv('../data/movie.csv')
```

```python
(movie['duration'] > 120).sum()
```

```python
movie['duration'].gt(120).sum()
```

2.전체 영화 중 상영 시간이 2시간이 넘는 영화의 비율을 확인하세요.

```python
movie['duration'].dropna().gt(120).mean()
```

- 누락값이 있으면 평균에 오차가 발생하므로 누락값을 먼저 처리

- value_counts(normalize=True)로 검증할 수 있다.

```python
movie['duration'].dropna().gt(120).value_counts(normalize=True)
```

## 다중 불리언 조건

2.영화 데이터셋을 로드하여 평점(imdb_score)이 8점 이상, 등급(content_rating)이 PG-13, 2000년~2009년 사이에 출시(titie_year)된 영화를 찾으세요.

```python
movie = pd.read_csv('../data/movie.csv', index_col='movie_title')
```

```python
crit1 = movie['imdb_score'] >= 8
```

```python
crit2 = movie['content_rating'] == 'PG-13'
```

```python
crit3 = (movie['title_year'] >= 2000) & (movie['title_year'] <= 2009)
```

```python
crit_all = crit1 & crit2 & crit3
```

```python
movie[crit_all].head()
```

```python
movie[crit_all].shape
```

## 불리언 인덱싱

2-1. 모든 조건에 부합되는 영화의 평점, 등급, 출시년도를 모아 별도의 DataFrame을 생성하세요.

```python
movie2 = movie[crit_all]
```

```python
movie3 = movie2[['imdb_score', 'content_rating', 'title_year']]
```
