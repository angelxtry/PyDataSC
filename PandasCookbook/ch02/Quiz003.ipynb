6.대학 데이터셋을 index를 INSTNM으로 로드하여 'UGDS_'로 시작하는 col만 추출하여 college_ugds DataFrame을 생성하세요.

```python
college = pd.read_csv('../data/college.csv', index_col='INSTNM')
```

```python
college_ugds = college.filter(like='UGDS_')
```

```python
college_ugds.head()
```

7.모든 cols이 누락값인 항목을 제외하여 college_ugds_drop DataFrame을 생성하세요.

```python
college_ugds_drop = college_ugds.dropna(how='all')
```

```python
college_ugds.shape
```

```python
college_ugds_drop.shape
```

8.college_ugds_drop DataFrame을 백인(UGDS_WHITE) 비율이 가장 높은 순으로 정렬하세요.

```python
college_ugds_drop.sort_values(by=['UGDS_WHITE'], ascending=False).head()
```

8.college_ugds_drop DataFrame을 아시안(UGDS_ASIAN) 비율이 가장 높은 대학 중 흑인 비율이 가장 낮은 순으로 정렬하세요.

```python
college_ugds_drop.sort_values(by=['UGDS_AIAN', 'UGDS_BLACK'], ascending=[False, True]).head()
```
